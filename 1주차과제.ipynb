### **■ 언어분야 : 텍스트 요약**

#### **1.페가수스(Google Brain)**
>&nbsp;&nbsp;&nbsp;&nbsp;구글 브레인 연구팀이 영국 임페리얼 칼리지 런던과 협업하여 AI 기반의 자동 텍스트 요약 시스템이다. 하나의 요약문 생성을 위해 여러 텍스트 부분들을 선택·조합하는 기존의 텍스트 요약 시스템보다 AI 기술을 이용해 새로운 단어를 생성해 각기 다른 부분들을 연결함으로써 언어학적으로 유려하고 일관성 있는 자연스러운 요약본을 만들어 낸다. 
<br> 
&nbsp;&nbsp;&nbsp;&nbsp;가장 뛰어난 성능을 보인 페가수스 모델은 5억 6,800만 개의 매개변수 또는 기존 데이터로부터 학습된 변수가 포함되어  최적화된 프로그램이고, 알고리즘을 학습시키기 위해 3억 5,000만 개의 텍스트 및 뉴스와 뉴스 유사 웹사이트에서 수집한 15억 개 기사 등을 통해 훈련되었다.
<br> 
&nbsp;&nbsp;&nbsp;&nbsp;페가수스 모델은 뉴스에서부터 과학, 특허, 법안, 이메일 등 다양한 영역의 텍스트 요약이 가능하다. 이 시스템이 일부 직종에서 시간 절감과 업무의 효율성 개선 효과를 가져올 것으로 기대하고 있다.

* **관련 논문명** : [PEGASUS: Pre-training with Extracted Gap-sentences for Abstractive Summarization](https://arxiv.org/abs/1912.08777)

<br>

#### **2.아골로(Agolo)**
>&nbsp;&nbsp;&nbsp;&nbsp;통상 AI 요약은 원문의 문장을 그대로 들여와 배열만 다시 하는 추출 방식과 요약문을 새롭게 다시 쓰는 초록 방식이 있다. 아골로 서비스는 초록과 추출을 혼용하는 하이브리드 방식을 구현하여 문장을 다루는 실력이 인간 못지않다는 평을 받고 있다.
<br> 
&nbsp;&nbsp;&nbsp;&nbsp;AP통신은 과거 에디터가 일일이 썼던 TV 뉴스 앵커용 요약문을 아골로의 AI에 맡기는 실험 프로젝트를 2018년부터 진행하여 에디터의 일손을 대거 아끼고 생산하는 요약문의 양도 크게 늘려 일거양득 효과를 기대하고 있다. 뉴욕타임스와 같은 언론사, 금융, 컨설팅, IT업계 등에 보고서 요약 서비스를 제공하고 있다.

* **Agolo 홈페이지 :**[ https://www.agolo.com/](https://www.agolo.com/)

<br>

#### **3.MoYa App(시스메틱)**
>&nbsp;&nbsp;&nbsp;&nbsp;모야(MoYa)는 AI 기반으로 뉴스를 매번 번거롭게 검색할 필요가 없이 구독자가 필요로 하는 맞춤 설정으로 실시간으로 필요한 뉴스와 정보를 제공받는 알림 앱이다. 여러 기능 중 CNN을 사용하여 23개의 카테고리로 텍스트를 분류하고 정확한 분류를 위해 주기적으로 학습을 진행하며, 매번 150만 건의 자료를 이용해 국문과 영문 모두 텍스트 요약과 키워드를 추출한다.

* **시스메틱 홈페이지 :**[ http://moya.ai/](http://moya.ai/)

### **■ 자율주행분야 : 자율주행 자동차**
&nbsp;&nbsp;&nbsp;&nbsp;자율주행차는 운전자 또는 승개의 조작없이 자동차 스스로 운행이 가능한 자동차로서, 자동차 스스로 사람의 인지, 판단, 제어 기능을 대체하여 운전한다. 운전자의 개입여부, 자동화 수준에 따라 6단계로 구분되며, Level 3 단계 이상은 일부 또는 완전 자율주행 기능을 수행한다. 자율주행차를 구성하는 여러가지 기술 중 판단기술은 기존 고가의 센서들을 사용하여 규칙기반 방식을 사용하였으나 이미지 인식 분야에 인공지능을 적용시키고, 이외에도 경로 최적화, 상황판단, 충돌예측, 돌발상황 대응 등에도 인공지능 기술을 활용하고 있다. 아직까지 윤리적 딜레마, 시스템 오류 등 문제점이 있지만 인공지능 기술과 하드웨어가 발전하면서 자율주행차도 완성도가 높아지고 있다.
